# 3.Load Models
In this notebook,we will need detectionmodel.When we download models,we need to initialize inference engine(IECore),and use read_network to read network architecture and weights from .xml and .bin files.Then,we load the network on the "CPU" using load_network.

```python
# Initialize inference engine

ie_core = Core()


def model_init(model_path: str):
    """
    Read the network and weights from file, load the
    model on the CPU and get input and output names of nodes

    :param: model: model architecture path *.xml
    :retuns:
            input_key: Input node network
            output_key: Output node network
            exec_net: Encoder model network
            net: Model network
    """
    # Read the network and corresponding weights from file
    model = ie_core.read_model(model=model_path)
    # compile the model for the CPU (you can use GPU or MYRIAD as well)
    compiled_model = ie_core.compile_model(model=model, device_name="CPU")
    # Get input and output names of nodes
    input_keys = compiled_model.input(0)
    output_keys = compiled_model.output(0)
    return input_keys, output_keys, compiled_model
```

### 3.1 Get attributes from model


```python
# Detection model initialization
input_key, output_keys, compiled_model = model_init(model_path)
# Get input size - Detection
height, width = list(input_key.shape)[2:]
```

```python
# Function for showing Raw Image
def plt_img(raw_image):
    """
    Use matplot to show image inline
    raw_image: input image

    :param: raw_image:image array
    """
    plt.figure(figsize=(10, 6))
    plt.axis("off")
    plt.imshow(raw_image)
```

####  3.2.Read and show a test image
From detection model input shape [1, 3, 480, 864],we need to resize the image size to 864x480 resolution.,and expand batch channel with expand_dims function.

```python
# Read a image
image_de = cv2.imread("../205-vision-background-removal/data/coco_cross.png")
# Resize to [3, 256, 256]
resized_image_de = cv2.resize(image_de, (width, height))
# Expand to [1, 3, 256, 256]
input_image_de = np.expand_dims(resized_image_de.transpose(2, 0, 1), 0)
# Show image
plt_img(raw_image=cv2.cvtColor(image_de, cv2.COLOR_BGR2RGB))
```

```python
# Create inference request
request = compiled_model.create_infer_request()
request.infer({input_key.any_name: input_image_de})
boxes = request.get_tensor("boxes").data
# Remove zero only boxes
boxes = boxes[~np.all(boxes == 0, axis=1)]
```

```python
# For each detection, the description has the format:
# [x_min, y_min, x_max, y_max, conf]
# Image passed here is in BGR format with changed width and height.
# To display it in colors expected by matplotlib we use cvtColor function


def result_to_image(bgr_img, resize_img, boxes, thold=0.3, conf_labels=True):
    # Define colors for boxes and descriptions
    colors = {"red": (255, 0, 0), "green": (0, 255, 0)}
    # Fetch image shapes to calculate ratio
    (real_y, real_x) = bgr_img.shape[:2]
    (resize_y, resize_x) = resize_img.shape[:2]
    ratio_x, ratio_y = real_x / resize_x, real_y / resize_y
    # Convert base image from bgr to rgb format
    rgb_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB)
    # Iterate through non-zero boxes
    for box in boxes:
        # Pick confidence factor from last place in array
        conf = box[-1]
        # thold = Threshold
        if conf > thold:
            # Convert float to int
            # Then multiply corner position of each box by x and y ratio
            # In case that bounding box is found at the top of the image,
            # we position upper box bar little lower
            # to make it visible on image
            (x_min, y_min, x_max, y_max) = [
                            int(max(corner_pos * ratio_y, 10))
                            if idx % 2
                            else int(corner_pos * ratio_x)
                            for idx, corner_pos in enumerate(box[:-1])
                                           ]
            # Draw box based on position,
            # parameters in rectangle function are:
            # image, start_point, end_point, color, thickness
            rgb_image = cv2.rectangle(rgb_img,
                                      (x_min, y_min), (x_max, y_max),
                                      colors["green"], 3)
            # Add text to image based on position and confidence
            # Parameters in text function are:
            # image, text, bottom-left_corner_textfield, font,
            # font_scale, color, thickness, line_type
            if conf_labels:
                rgb_image = cv2.putText(
                    rgb_image,
                    f"{conf:.2f}",
                    (x_min, y_min - 10),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    0.8,
                    colors["red"],
                    1,
                    cv2.LINE_AA,
                )
    return rgb_image
```

```python
# Detected Image
plt.figure(figsize=(10, 6))
plt.axis("on")
plt.imshow(result_to_image(
                            image_de,
                            resized_image_de, boxes, conf_labels=True));
```
